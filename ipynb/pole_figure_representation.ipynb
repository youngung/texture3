```python
fntex='zr.tex'
dat=np.loadtxt(fntex,skiprows=4)

## new data is in the dimension of 5210 x 4
newdat=np.zeros(dat.shape)
for ig in range(len(dat)):
    ## for each row (=each discrete orientation)
    phi1,phi,phi2,wgt=dat[ig]

    ## transformation matrix that converts from SA to CA
    Rg=euler(phi1,phi,phi2,echo=False)

    ## new transformation matrix (new R = Rg . old R)
    newR=np.zeros((3,3))
    for i in range(3):
        for j in range(3):
            for k in range(3):
                newR[i,j]=newR[i,j]+Rg[i,k]*R[k,j]
                
    ## convert new R to Euler angles.
    phi1,phi,phi2=euler(a=newR)
    
    ## store new Euler angles to data
    newdat[ig,:]=phi1,phi,phi2,wgt
```

```python
fntex='zr.tex'
mypf=upf.polefigure(filename=fntex,csym='hexag',cdim=[1,1,1.594],cang=[90,90,120])
fig=mypf.pf_new(mode='fill',poles=[[0,0,0,2],[1,0,-1,0]],
                ix='   1//TD',iy='2//ND',
                levels=[0.5,0.7,1,1.4,2,3,4,6])
fig.savefig('/tmp/init_zr.pdf')
mypf=upf.polefigure(grains=newdat,csym='hexag',cdim=[1,1,1.594],cang=[90,90,120])
fig=mypf.pf_new(mode='fill',poles=[[0,0,0,2],[1,0,-1,0]],
                ix='   1//RD',iy='2//TD',
                levels=[0.5,0.7,1,1.4,2,3,4,6])
fig.savefig('/tmp/init_zr_after_rotation.pdf')
```
