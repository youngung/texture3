- Miller index {h,k,l} to unit vectors $n$ and $b$

```python
h,k,l=1,1,1
u,v,w=1,-1,0
n=np.array([h,k,l])
b=np.array([u,v,w])
n=n/(n**2).sum()
b=b/(b**2).sum()
```

```python
schmid=np.zeros((3,3))
for i in range(3):
    for j in range(3):
        schmid[i,j]=(n[i]+b[j])/2.
```

$$\dot{\gamma}/\dot{\gamma}_0=(\tau/\tau_0)^n$$

```python
gamd0=0.1
n=20.
tau0=1.
tau=np.linspace(0,1.4,200)

fig=plt.figure(figsize=(5,4));ax=fig.add_subplot(111)
ns=[1,2,5,10,20,50,10000]
for i in range(len(ns)):
    gamd=gamd0*(tau/tau0)**ns[i]

    ax.plot(tau/tau0,gamd/gamd0,label='$n=%i$'%ns[i])
ax.set_xlabel(r'$\tau/\tau_0$')
ax.set_ylabel(r'$\dot{\gamma}/\dot{\gamma}_0$')
ax.set_ylim(-0.1,1.1)
ax.set_xlim(0,1.1)
ax.plot([1,1],[0,1.4],c='k',ls='--')
ax.set_title(r'$\dot{\gamma}/\dot{\gamma}_0=(\tau/\tau_0)^n$')
ax.legend()
fig.tight_layout()
```
